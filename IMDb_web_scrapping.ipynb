```python
# remove "\N" from starYear column
df_movies = df_movies.loc[df_movies["startYear"].isin(["\\N"]) == False,]
```

```python
# change the dtype of startYear from str to numeric
df_movies["startYear"] = pd.to_numeric(df_movies["startYear"])
```

```python
#filters
titleType = df_movies["titleType"]=="movie"
startYear = df_movies["startYear"]>2015
# originalTitle = df_movies["originalTitle"].str.contains('Top Gun')

#apply the filter to resize data set.
df_movies = df_movies.loc[titleType & startYear,]
```
